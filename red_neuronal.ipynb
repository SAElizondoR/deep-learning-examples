```python
import os
os.environ["CUDA_VISIBLE_DEVICES"] = "-1"

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

import tensorflow as tf

from keras.models import Sequential
from keras.layers import Dense, Input
```

```python
print("Num GPUs Available: ", len(tf.config.list_physical_devices('GPU')))
```

```python
# Función para cargar, dividir y estandarizar el conjunto de datos
def cargar_datos(ruta, test_size=0.2, columnas_x=None):
    ERDB = pd.read_csv(ruta)
        
    # Definir columnas por defecto si no se proporcionan
    columnas_x = columnas_x or [
        "Gender", "Desease", "Education Level", "Marital Status",
        "Monthly Income", "Employee/Employer", "Dependants",
        "Unemployment Rate", "Stock Market", "Credit Score",
        "Government Bonds Return"
    ]

    # Dividir el conjunto de datos
    train, test = train_test_split(ERDB, test_size=test_size, random_state=42)
    
    # Dividir en característics y etiquetas
    train_x, train_y = train[columnas_x], train[["Retire Before 65 Years Old"]]
    test_x, test_y = test[columnas_x], test[["Retire Before 65 Years Old"]]
    
    # Estandarizar de los datos
    scaler = StandardScaler()
    train_xs = scaler.fit_transform(train_x)
    test_xs = scaler.transform(test_x)
    
    return train_xs, train_y, test_xs, test_y, scaler
```

```python
# Ejecutar el flujo de datos
ruta_datos = '../EarlyRetirementPrediction.csv'
train_xs, train_y, test_xs, test_y, scaler = cargar_datos(ruta_datos)
```

```python
# Mostrar primeras filas para verificación
display(pd.DataFrame(train_xs, columns=scaler.feature_names_in_).head())
```

```python
# Verificar longitud de los datos de entrenamiento
print(f'Tamaño de datos de entrenamiento: {len(train_xs)}')
```

```python
# DEFINIR RED NEURONAL
model = Sequential([
    Input(shape=(11,)),
    Dense(5, activation='relu'),
    Dense(3, activation='sigmoid'),
    Dense(1, activation='sigmoid')
])
```

```python
# DEFINIR LA FUNCIÓN DE PÉRDIDA Y EL OPTIMIZADOR
model.compile(loss='binary_crossentropy', optimizer='adam',
    metrics=['accuracy'])
```

```python
model.fit(train_xs, train_y, epochs=150, batch_size=10)
```
